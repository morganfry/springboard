## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#load JSON as dataframe
json_df = pd.read_json('data/world_bank_projects.json')
```

```python
#examine head of dataframe
json_df.head()
```

```python
json_df.columns #examine the columns for the most useful one use
```

<p><b>Each row of the data represents a single project. To find which countries have how many projects we can count the occurrences of the country names (shortened names for the sake of readability).</b></p>

```python
counts=json_df.countryshortname.value_counts()
top_10=counts.head(10)
print("The top 10 countries with the most projects are: \n")
print(top_10)

#plot the top 10 results
top_10_r=top_10.iloc[::-1] #reverse for barh display
top_10_r.plot.barh()
```

<p><b>Next, we'll determine the 10 most common project themes.</b></p>

```python
#first examine the relevant column 'mjtheme_namecode'
namecodes=json_df.mjtheme_namecode
namecodes.head()
```

<p><b>We can see that this has nested JSON data, so it needs flattening into a separate dataframe.</b></p>

```python
#load json as string
json_data=json.load((open('data/world_bank_projects.json')))

#normalize the json into a dataframe
namecode_df=json_normalize(json_data,'mjtheme_namecode')
namecode_df.head(15)
```

<p><b>This has missing data so we need to fill it in.</b></p>

```python
#first make a dataframe of codes and their names
codes_df=namecode_df[namecode_df['name'] != ""].drop_duplicates()
codes_df
```

```python
#then fill in the missing data by merging the two dataframes
namecode_merged=codes_df.merge(right=namecode_df, on='code', how='left')
namecode_merged
```

```python
#rename relevant columns
namecode_merged.rename(columns={'code':'code','name_x':'Project_Theme'}, inplace=True)

#bounce relevant column to new dataframe, using same value_counts method as before for top 10 
top_10_prj=namecode_merged.Project_Theme.value_counts().head(10)

#display top 10 results
print("The top 10 project themes are:\n")
print(top_10_prj)

#plot the top 10 results
top_10_prj_r=top_10_prj.iloc[::-1] #reverse for barh display
top_10_prj_r.plot.barh(clip_on=True)
```
